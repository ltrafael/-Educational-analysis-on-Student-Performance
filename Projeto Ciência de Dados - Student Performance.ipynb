**Base de dados codificada através de Label Enconding ou Codificação por Rotulação:**

````python
def LE(df):
    """
    função que transforma variáveis nominais em tipo categoria e aplica codificação por rotulação, sem ordem específica
    """
    for i in df.columns:
        if df[i].dtype == 'object':
            df[i] = df[i].astype('category')
            df[i] = df[i].cat.codes
````

A melhor solução varia para cada atributo. Para atributos nominais onde não há uma ordem intrínseca dentre as possíveis classes One-Hot-Encoding é ideal. Para atributos em que é possível atribuir uma ordem, codificação por rotulação é mais indicado por representar esta ordem.

**Padronização:**

````python
# biblioteca sci-kit learn
from sklearn.preprocessing import StandardScaler

# copia base original
df_esc1 = df.copy()
# cria escalonador
standard_scaler = StandardScaler()
# lista com as colunas a serem tarnsformadas
cols_esc = ['age', 'failures', 'absences', 'G1', 'G2', 'G3']

# aplica padronizador para cada coluna separadamente
df_esc1[cols_esc] = standard_scaler.fit_transform(df_esc1[cols_esc])

df_esc1[cols_esc].head()
````

Padronização transforma cada atributo de modo que a média dos valores seja 0 e o desvio padrão seja 1. Cada instância é transformada de acordo com a fórmula a seguir:
$$ z = \frac{x - u}{\sigma} $$

Sendo $z$ o novo valor a ser gerado, $x$ o valor atual da instância, $u$ a média do atributo e $\sigma$ o desvio padrão do atributo. Podemos verificar para cada atributo se o processo foi calculado corretamente calculando estas duas estatísticas.

````python
for col in df_esc1[cols_esc]:
    print (f'{col:10}', '\tmean:', f'{round(df_esc1[col].mean(), 2):4}', '\tstd:', round(df_esc1[col].std(), 2))
````

**Escalonamento pelo vetor de comprimento unitário:**

````python
from sklearn.preprocessing import Normalizer

# copia base original
df_esc6 = df.copy()
# cria instância do normalizador
Norm = Normalizer()
# aplica normalizador para atributos
df_esc6[cols_esc] = Normalizer().fit_transform(df_esc6[cols_esc])

df_esc6[cols_esc].head()
````

Escalonamento pelo vetor de comprimento unitário, ou normalização, é feita pela classe Normalizer do sklearn e tem como parâmetro padrão *norm='l2*. Isto significa que está utilizando a distância euclidiana (L2) para calcular o padrão dos novos valores, que é o método mais utilizado. Diferentemente das outras transformações até agora, a normalização é feita por linha (instância), e não por coluna  (atributo). O cálculo segue a fórmula a seguir:

$$ valor_{novo} = \frac{valor_{atual}}{L} $$

Sendo $L$ a distância calculada com todos os $n$ valores $i$ da instância. $L1$ representa a distância de Manhattan e $L2$ representa a distância euclidiana.

$$ L1_n = \sum_{i=1}^{n} |i_1| + |i_2| \,+\, ...\, + \, |i_n| $$

$$ L2_n = \sum_{i=1}^{n} \sqrt{i_1^2 + i_2^2 \,+\, ...\, +\, i_n^2} $$ 

Podemos conferir a transformação aplicando a fórmula de $L2_n$ nos valores $i$ de cada instância.

````python
df_esc6_test = df_esc6[cols_esc].copy()
df_esc6_test['L2'] = df_esc6_test.apply(lambda x: x.age**2 + x.failures**2 + x.absences**2 + x.G1**2 + x.G2**2 + x.G3**2,
                                       axis=1)
df_esc6_test.head()
````

**Imputação usando o final da distribuição;**

Pode-se usar imputação de três desvios padrão acima da média para distribuições normais ou imputação por regra interquartil. A regra interquartil atribui aos valores faltantes o limite IQR superior:

$$ Limite\; IQR\; superior = 3^{o} Quartil + (IQR * 1.5) $$

Sendo $ IQR = 3^{o} Quartil - 1^{o} Quartil $. Este método também não é aplicado ao atributo *reason*.

````python
def replace_limit(column):
    """
    função que retorna o limite superior de uma coluna
    """
    IQR = column.quantile(0.75) - column.quantile(0.25)
    limite_superior = column.quantile(0.75) + (IQR * 1.5)
    
    return limite_superior
````

````python
# copia base com valores faltantes
df_na_dist = df_na.copy()

# completa valores faltantes de cada coluna por 
df_na_dist.fillna({'age': replace_limit(df_na_dist['age']), 
                   'studytime': replace_limit(df_na_dist['studytime']), 
                   'absences': replace_limit(df_na_dist['absences'])}, inplace=True)

df_na_dist.head()
````

## 4. Análise de Outliers

Um dos métodos existentes de identificação de outliers é o método padrão que a biblioteca *seaborn* utiliza para plotar os gráficos de caixa, ou boxplot. Outliers são valores considerados além de um certo intervalo, com base na faixa interquartil (IQR) e no primeiro e terceiro quartis. A fórmula é a mesma implementada para a imputação usando o final da distribuição, para o limite superior, e para o limite inferior é a seguinte:

$$ Limite\; inferior = 1^{o} Quartil - (IQR * 1.5) $$

Todos os dados abaixo do limite inferior também são considerados outliers. Este cálculo é feito de maneira independente para cada atributo.

Outliers são existentes apenas em atributos com valores numéricos, seja este valor ordinal ou contínuo, intervalar ou razão.

````python
# lista com colunas numéricas
cols_num = df.select_dtypes('int64').columns

# há 16 atributos numéricos/ordinais na base df, todos são do tipo int64
cols_num.nunique()
````

````python
plt.figure(figsize=(16, 5))
sns.boxplot(data=df[cols_num]);
````

````python
# identificando outliers de forma automática
# a função replace_limit criada anteriormente calcula o limite superior de um atributo
# a função foi modificada para calcular também o limite inferior

def replace_limit(column):
    """
    função que retorna o limite superior e inferior de uma coluna
    """
    IQR = abs(column.quantile(0.75) - column.quantile(0.25))
    limite_superior = column.quantile(0.75) + (IQR * 1.5)
    limite_inferior = column.quantile(0.25) - (IQR * 1.5)
    
    return limite_superior, limite_inferior
````

````python
# nova função para contagem de outliers
def count_outliers(df, cols_num):
    """
    usa a função replace_limit para verificar outliers e usa a variável aux para contagem de colunas
    """
    
    # auxiliar para contar colunas com outliers
    aux = 0

    # loop para cada coluna numérica
    for coluna in cols_num:
        lim_sup, lim_inf = replace_limit(df[coluna]) # pega valores do limite superior e inferior de cada coluna
        outliers = df[(df[coluna] > lim_sup) | (df[coluna] < lim_inf)][coluna].count() # conta quantas ocorrências de outliers
        if outliers > 0:
            aux += 1 # se tiver outliers, acresce aux em 1

        print (f'Coluna: {coluna:10}   número de outliers: {outliers:3}')

    print (f'\nTotal de colunas com outliers: {aux}')
    
count_outliers(df, cols_num)
````

• **Gere uma nova base em que não haja outliers;**

````python
# copia base original
df_no_out = df.copy()

# refaz loop anterior, salvando em um set os valores dos índices com outliers
# foi utilizado set para juntar todos os índices sem repetição
set_outlier = set()
for coluna in cols_num:
    lim_sup, lim_inf = replace_limit(df_no_out[coluna])
    outliers = df_no_out[(df_no_out[coluna] > lim_sup) | (df_no_out[coluna] < lim_inf)].index # índices dos outliers
    set_outlier.update(list(outliers)) # adiciona lista com índices no set
    
len(set_outlier)
````

• **Gere uma nova base fazendo o recorte de outliers usando Média e Desvio Padrão.**

Este método é similar ao método anterior, que faz o recorte de outliers usando o IQR, apenas agora os limites serão calculados da seguinte forma:

$$ Limite\; superior = u + (\sigma * 3) $$

$$ Limite\; inferior = u - (\sigma * 3) $$

Sendo $u$ a média do atributo e $\sigma$ o desvio padrão.

````python
# copia base original
df_out_std = df.copy()

# loop para cada uma das 16 colunas numéricas, substituindo outliers pelo limite superior e inferior
for coluna in cols_num:
    lim_sup, lim_inf = replace_limit(df_out_std[coluna])
    df_out_std[coluna].replace(df_out_std[df_out_std[coluna] > lim_sup][coluna], lim_sup, inplace=True)
    df_out_std[coluna].replace(df_out_std[df_out_std[coluna] < lim_inf][coluna], lim_inf, inplace=True)
````

## 7. Realize seleção dos atributos existentes na base de dados (tanto nominais quanto numéricos):

**• Escolha um método do tipo filtro para seleção;**

Métodos de filtro para seleção de atributos são métodos supervisionados que avaliam a importância de cada atributo para a variável *target*. O método de informação mútua foi aplicado, e depois separadamente para atributos numéricos, o algoritmo *CFS* (*Correlation-based feature selection*) foi aplicado. 

Os atributos *G1* e *G2* foram excluídos da análise para evitar multicolinearidade.

````python
# bibliotecas para fazer a seleção de atributos
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression # CFS
from sklearn.feature_selection import mutual_info_regression # informação mútua
````

````python
# separa atributos numéricos
# há apenas 5 atributos numéricos: age, failures, absences, G1, G2 e G3. Sendo G3 o target
cols_num = ['age', 'failures', 'absences', 'G1', 'G2']
````

````python
# aplica método de informação mútua

# divide o target da base
X = df.drop(['G1', 'G2', 'G3'], axis=1).copy()
y = df['G3'].copy()

# divide dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

# aplica one-hot-encoding nos dados de treino e teste
X_train_dummie = pd.get_dummies(X_train, drop_first=True)
X_test_dummie = pd.get_dummies(X_test, drop_first=True)

# cria instância, encaixa com dados de treino e transforma X_train_dummie e X_test_dummie
mir = SelectKBest(score_func=mutual_info_regression, k='all')
mir.fit(X_train_dummie, y_train)
X_train_mir = mir.transform(X_train_dummie)
X_test_mir = mir.transform(X_test_dummie)
````

````python
# aplica método CFS para seleção de atributos numéricos apenas

# divide o target da base
X = df[cols_num].copy()
y = df['G3'].copy()

# divide dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)

# cria instância, encaixa com dados de treino e transforma X_train e X_test
Kbest = SelectKBest(f_regression, k='all')
Kbest.fit(X_train, y_train)
X_train_k = Kbest.transform(X_train)
X_test_k = Kbest.transform(X_test)
````
